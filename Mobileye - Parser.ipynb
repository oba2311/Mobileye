```python
decimaldata = []
for i in range(21):
    df['Decimal_ %s' % i] = df[i].apply(lambda x: int(str(x),16))
#     print (df[i])
    decimaldata.append(df['Decimal_ %s' % i])

decimaldata = pd.DataFrame(decimaldata)
decimaldata = decimaldata.T
decimaldata.head()
```

```python
# FCS:
col_list= list(decimaldata)
OmmitIndices = [0,21,20]
summCols = [i for i in col_list if col_list.index(i) not in OmmitIndices]
df['FCS'] = df[summCols].sum(axis=1)
df['FCS'].head()
# df['FCSFinal'] = df['FCS'].apply(lambda x: (255-x)%256)
```

```python
df['FCSFinal'] = df['FCS'].apply(lambda x: 255-x%256)
df['FCSFinal'].head()
```

```python
df['isError'] = np.where(decimaldata['Decimal_ 20']==df['FCSFinal'] , 1, 0)
plt.plot(df['isError'])
```
