```python
import seaborn as sns
sns.set(style="ticks", color_codes=True)
```

```python
import pandas as pd
from matplotlib import  pyplot as plt
data = pd.read_csv('./Test_Data (2).txt', delimiter="\t", header='infer')
data.head()
```

```python
data['severity'] = data['severity'].str.capitalize()
```

```python
data["severity"]=data["severity"].astype('category')
data["severity_cat"] = data["severity"].cat.codes
```

```python
sns.catplot(x="Driver_name", y="severity", data=data);
```

#### Q1: Looking at the data, it seems like Ringo is quite pasimistic - perhaps he should start using the 'Minor' option as well.

```python
data["road_type"]=data["road_type"].astype('str')
```

```python
# ! pip install statsmodels
from statsmodels.graphics.mosaicplot import mosaic
plt.rcParams['font.size'] = 16.0
mosaic(data, ['road_type', 'severity']);
```

```python
cross = pd.crosstab(data["road_type"], data["severity"])
cross.plot.bar(stacked=True)
plt.legend(title='severity stacked bars',loc='best', bbox_to_anchor=(-0.01, .93, -.1, 0.1)
)
plt.show()
```

```python
cross = pd.crosstab(data["Country"], data["severity"])
cross.plot.bar(stacked=True)
plt.legend(title='severity stacked bars',loc='best', bbox_to_anchor=(-0.01, .93, -.1, 0.1)
)
plt.show()
```

### We see that Italy is the worst.
